```python
x=df.drop(["CLASS"],axis=1)
```

```python
x
```

```python
y=df.loc[:,"CLASS"]
```

```python
y
```

```python
from sklearn.cluster import KMeans
KMeans_model=KMeans(n_clusters=3)
```

```python
KMeans_model.fit(x,y)
```

```python
test_data=pd.DataFrame()
test_data["VAR1"]=[0.906]
test_data["VAR2"]=[0.606]
```

```python
test_data
```

```python
y_pred=KMeans_model.predict(test_data)
```
